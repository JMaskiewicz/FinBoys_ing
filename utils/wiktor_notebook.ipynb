# Function to calculate EMA for a row
def calculate_ema(row, N):
    k = 2 / (N + 1)
    ema = row[0]  # Initialize EMA with the first value in the row
    for value in row[1:]:
        ema = (value * k) + (ema * (1 - k))
    return ema
