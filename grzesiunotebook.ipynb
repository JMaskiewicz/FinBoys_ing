```python
import pandas as pd

file_path = "C:\\Users\\GS\Desktop\\ING\\FinBoys_ing\\Data\\in_time.csv"
train = pd.read_csv(file_path)
```

```python
columns = ["Birth_date", "Contract_origination_date", "Contract_end_date", "Oldest_account_date"]

ref = pd.to_datetime(train['Ref_month'], format='%m-%Y')
for col in columns:
    d = pd.to_datetime(train[col], format='%d-%m-%Y')
    train[col + '_diff'] = abs(ref - d).dt.days
    print(train[col + '_diff'])
```
